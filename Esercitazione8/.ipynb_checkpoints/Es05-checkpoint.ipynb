```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
with open("100.txt") as dat100:
    dati_100 = np.fromstring(dat100.read(), sep=" ")
with open("isto.txt") as isto:
    isto = np.fromstring(isto.read(), sep=" ")
```

```python
with open("errori_100.txt") as errori_100:
    dati_err100 = np.fromstring(errori_100.read(), sep=" ")
```

```python
plt.errorbar(np.arange(len(dati_100)), dati_100 , yerr=dati_err100)
plt.xlabel('numero di blocchi')
plt.ylabel('<raggio>')
plt.title('stato 100')
plt.grid(True)
plt.show()
```

```python
bins = np.linspace(-3, 3, 100)
plt.hist(isto, bins, alpha=0.5, density=True)

plt.xlabel('x [raggi di Bohr]')
plt.ylabel('probabilità')
plt.title('distribuzione dei raggi')
plt.grid(True)
plt.show()
```
