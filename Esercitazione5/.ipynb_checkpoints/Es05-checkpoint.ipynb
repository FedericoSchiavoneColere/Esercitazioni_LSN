```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
with open("./Unif/100.txt") as U_dat100:
    U_dati_100 = np.fromstring(U_dat100.read(), sep=" ")
with open("./Unif/210.txt") as U_dat210:
    U_dati_210 = np.fromstring(U_dat210.read(), sep=" ")
```

```python
with open("./Unif/errori_100.txt") as U_errori_100:
    U_dati_err100 = np.fromstring(U_errori_100.read(), sep=" ")
with open("./Unif/errori_210.txt") as U_errori_210:
    U_dati_err210 = np.fromstring(U_errori_210.read(), sep=" ")
```

```python
with open("./Gauss/100.txt") as G_dat100:
    G_dati_100 = np.fromstring(G_dat100.read(), sep=" ")
with open("./Gauss/210.txt") as G_dat210:
    G_dati_210 = np.fromstring(G_dat210.read(), sep=" ")
```

```python
with open("./Gauss/errori_100.txt") as G_errori_100:
    G_dati_err100 = np.fromstring(G_errori_100.read(), sep=" ")
with open("./Gauss/errori_210.txt") as G_errori_210:
    G_dati_err210 = np.fromstring(G_errori_210.read(), sep=" ")
```

```python
fig, ax = plt.subplots()
ax.errorbar(np.arange(len(U_dati_100)), U_dati_100 , yerr=U_dati_err100)
ax.errorbar(np.arange(len(G_dati_100)), G_dati_100 , yerr=G_dati_err100)
plt.xlabel('numero di blocchi')
plt.ylabel('<raggio>')
plt.title('stato 100')
plt.legend(['Uniforme', 'Gaussiana'])
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.errorbar(np.arange(len(U_dati_210)), U_dati_210 , yerr=U_dati_err210)
ax.errorbar(np.arange(len(G_dati_210)), G_dati_210 , yerr=G_dati_err210)
plt.xlabel('numero di blocchi')
plt.ylabel('<raggio>')
plt.title('stato 210')
plt.legend(['Uniforme', 'Gaussiana'])
plt.grid(True)
plt.show()
```
