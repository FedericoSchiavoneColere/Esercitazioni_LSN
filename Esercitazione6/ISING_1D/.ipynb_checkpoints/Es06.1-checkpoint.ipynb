```python
import os
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
media_column = []
err_column = []

with open('output/output_1.00_ene.txt', 'r') as file:
    for line in file:
        columns = line.split()
        
        if len(columns) >= 4:  # Ensures there are at least four columns
            second_value = columns[2].strip()
            fourth_value = columns[3].strip()

            if second_value and fourth_value:  # Skips empty strings
                media_column.append(float(second_value))
                err_column.append(float(fourth_value))

plt.title("Energia a T=1K")
plt.xlabel("Numero di blocchi")
plt.ylabel("U/N")
plt.errorbar(np.arange(len(media_column)), media_column , yerr=err_column)
plt.grid(True)
plt.show()
```

```python
media_column = []
err_column = []

with open('output/output_1.00_chi.txt', 'r') as file:
    for line in file:
        columns = line.split()
        
        if len(columns) >= 4:  # Ensures there are at least four columns
            second_value = columns[2].strip()
            fourth_value = columns[3].strip()

            if second_value and fourth_value:  # Skips empty strings
                media_column.append(float(second_value))
                err_column.append(float(fourth_value))

plt.title("Suscettibilit magnetica a T=1K")
plt.xlabel("Numero di blocchi")
plt.ylabel("$\chi$")    
plt.errorbar(np.arange(len(media_column)), media_column , yerr=err_column)
plt.grid(True)
plt.show()
```

```python
# Define your pre-defined function
def function_ene(T):
    beta = 1/T
    J = 1.0
    Ns = 50
    th = np.tanh(J/T)
    thN = th**Ns
    ch = 1/th
    e = -J*(th + ch*thN)/(1 + thN)
    return e

# Path to the directory containing your files
directory = './output/'

# Get a list of all the files in the directory
files = os.listdir(directory)

# Arrays to store the mean values, error bars, and T values
mean_values = []
error_bars = []
T_values = []

# Iterate over each file
for file_name in files:
    # Check if the file is a text file
    if file_name.endswith('ene.txt'):
        file_path = os.path.join(directory, file_name)

        # Extract the T value from the file name
        match = re.search(r'output_(\d+\.\d+)_ene.txt', file_name)
        if match:
            T_value = float(match.group(1))
        else:
            continue

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Extract the second-to-last row and split it by spaces
        second_last_row = lines[-2].split()

        # Extract the last two elements (mean value and error bar)
        mean_value = float(second_last_row[-2])
        error_bar = float(second_last_row[-1])

        # Append the values to the arrays
        mean_values.append(mean_value)
        error_bars.append(error_bar)
        T_values.append(T_value)

# Generate x values for plotting the function
x = np.linspace(0.2, 8, 100)

sorted_data = sorted(zip(T_values, mean_values, error_bars))
T_values, mean_values, error_bars = zip(*sorted_data)

# Plot the function
plt.plot(x, function_ene(x), label='Function')

# Plot the mean values with error bars
for mean, err, T in zip(mean_values, error_bars, T_values):
    plt.errorbar(T, mean, yerr=err, fmt='o', label=f'T = {T:.2f}')

plt.xlabel('T')
plt.ylabel('U/N')
plt.legend(loc='upper right', bbox_to_anchor=(1.8, 1),ncol=3)
plt.title('Energia(T)')
plt.show()
```

```python
# Define your pre-defined function
def function_heat(T):
    beta = 1/T
    J = 1.0
    Ns = 50
    th = np.tanh(J/T)
    thN = th**Ns
    ch = 1/th
    heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
    return heat

# Path to the directory containing your files
directory = './output/'

# Get a list of all the files in the directory
files = os.listdir(directory)

# Arrays to store the mean values, error bars, and T values
mean_values = []
error_bars = []
T_values = []

# Iterate over each file
for file_name in files:
    # Check if the file is a text file
    if file_name.endswith('heat.txt'):
        file_path = os.path.join(directory, file_name)

        # Extract the T value from the file name
        match = re.search(r'output_(\d+\.\d+)_heat.txt', file_name)
        if match:
            T_value = float(match.group(1))
        else:
            continue

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Extract the last row and split it by spaces
        last_row = lines[-1].split()

        # Extract the last two elements (mean value and error bar)
        mean_value = float(last_row[-2])
        error_bar = float(last_row[-1])

        # Append the values to the arrays
        mean_values.append(mean_value)
        error_bars.append(error_bar)
        T_values.append(T_value)


# Generate x values for plotting the function
x= np.linspace(0.1, 8, 100)

sorted_data = sorted(zip(T_values, mean_values, error_bars))
T_values, mean_values, error_bars = zip(*sorted_data)


# Plot the function
plt.plot(x, function_heat(x), label='Function')

for mean, err, T in zip(mean_values, error_bars, T_values):
    plt.errorbar(T, mean, yerr=err, fmt='o', label=f'T = {T:.2f}')

plt.xlabel('T')
plt.ylabel('C')
plt.legend(loc='upper right', bbox_to_anchor=(1.8, 1),ncol=3)
plt.title('Calore specifico(T)')
plt.show()
```

```python
# Define your pre-defined function
def function_chi(T):
    beta = 1/T
    J = 1.0
    Ns = 50
    th = np.tanh(J/T)
    thN = th**Ns
    X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
    return X

# Path to the directory containing your files
directory = './output/'

# Get a list of all the files in the directory
files = os.listdir(directory)

# Arrays to store the mean values, error bars, and T values
mean_values = []
error_bars = []
T_values = []

# Iterate over each file
for file_name in files:
    # Check if the file is a text file
    if file_name.endswith('chi.txt'):
        file_path = os.path.join(directory, file_name)

        
        # Extract the T value from the file name
        match = re.search(r'output_(\d+\.\d+)_chi.txt', file_name)
        if match:
            T_value = float(match.group(1))
        else:
            continue

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Extract the last row and split it by spaces
        last_row = lines[-1].split()

        # Extract the last two elements (mean value and error bar)
        mean_value = float(last_row[-2])
        error_bar = float(last_row[-1])

        # Append the values to the arrays
        mean_values.append(mean_value)
        error_bars.append(error_bar)
        T_values.append(T_value)

# Generate x values for plotting the function
x = np.linspace(0.1, 8, 100)

sorted_data = sorted(zip(T_values, mean_values, error_bars))
T_values, mean_values, error_bars = zip(*sorted_data)


# Plot the function
plt.plot(x, function_chi(x), label='Function')

for mean, err, T in zip(mean_values, error_bars, T_values):
    plt.errorbar(T, mean, yerr=err, fmt='o', label=f'T = {T:.2f}')

plt.xlabel('T')
plt.ylabel('$\chi$')
plt.legend(loc='upper right', bbox_to_anchor=(1.8, 1),ncol=3)
plt.title('Suscettibilit magnetica(T)')
plt.show()
```

```python
# Define your pre-defined function

#accertarsi che i dati iniziali in input.txt prevedano h=0.02


def function_mag(T):
    h=0.02 #external field
    b = 1/T
    J = 1.0
    Ns = 50
    th = np.tanh(J/T)
    thN = th**Ns
    ch = 1/th
    l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    Z = l1**Ns + l2**Ns
    M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
    return M

# Path to the directory containing your files
directory = './output/'

# Get a list of all the files in the directory
files = os.listdir(directory)

# Arrays to store the mean values, error bars, and T values
mean_values = []
error_bars = []
T_values = []

# Iterate over each file
for file_name in files:
    # Check if the file is a text file
    if file_name.endswith('mag.txt'):
        file_path = os.path.join(directory, file_name)

        # Extract the T value from the file name
        match = re.search(r'output_(\d+\.\d+)_mag.txt', file_name)
        if match:
            T_value = float(match.group(1))
        else:
            continue

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Extract the last row and split it by spaces
        last_row = lines[-1].split()

        # Extract the last two elements (mean value and error bar)
        mean_value = float(last_row[-2])
        error_bar = float(last_row[-1])

        # Append the values to the arrays
        mean_values.append(mean_value)
        error_bars.append(error_bar)
        T_values.append(T_value)

# Generate x values for plotting the function
x = np.linspace(0.1, 8, 100)

sorted_data = sorted(zip(T_values, mean_values, error_bars))
T_values, mean_values, error_bars = zip(*sorted_data)

# Plot the function
plt.plot(x, function_mag(x), label='Function')

for mean, err, T in zip(mean_values, error_bars, T_values):
    plt.errorbar(T, mean, yerr=err, fmt='o', label=f'T = {T:.2f}')

plt.xlabel('T')
plt.ylabel('M')
plt.legend(loc='upper right', bbox_to_anchor=(1.8, 1),ncol=3)
plt.title('Magnetizzazione (T, h=0.02)')
plt.show()


```
