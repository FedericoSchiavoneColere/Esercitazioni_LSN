```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as ss
```

```python
with open("int_unif.txt") as int_unif:
    dati_int_unif = np.fromstring(int_unif.read(), sep=" ")
with open("err_unif.txt") as err_unif:
    dati_err_unif = np.fromstring(err_unif.read(), sep=" ")
```

```python
plt.errorbar(np.arange(len(dati_int_unif)), dati_int_unif,yerr=dati_err_unif)
plt.xlabel('numero cicli')
plt.ylabel('valore integrale')
plt.grid(True)
#plt.ylim(-1,1)
plt.title('Estrazione Uniforme')
plt.show()
```

```python
with open("int_imp.txt") as int_imp:
    dati_int_imp = np.fromstring(int_imp.read(), sep=" ")
with open("err_imp.txt") as err_imp:
    dati_err_imp = np.fromstring(err_imp.read(), sep=" ")
```

```python
plt.errorbar(np.arange(len(dati_int_imp)), dati_int_imp, yerr=dati_err_imp)
plt.xlabel('numero cicli')
plt.ylabel('valore integrale')
plt.title('Estrazione Gaussiana')
plt.grid(True)
plt.show()

```
