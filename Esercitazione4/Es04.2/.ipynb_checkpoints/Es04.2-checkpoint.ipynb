```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
media_column = []
err_column = []

with open('./gas/output_temp.dat', 'r') as file:
    for line in file:
        columns = line.split()
        
        if len(columns) >= 4:  # Ensures there are at least four columns
            second_value = columns[2].strip()
            fourth_value = columns[3].strip()

            if second_value and fourth_value:  # Skips empty strings
                media_column.append(float(second_value))
                err_column.append(float(fourth_value))
```

```python
plt.errorbar(np.arange(len(media_column)), media_column , yerr=err_column)
plt.title('Temperatura - Gas', fontsize=17, fontweight="bold")
plt.ylabel('$T^*$', fontsize=15)
plt.xlabel('Step number', fontsize=15)
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.arange(1, N+1)    # use the ordering from 1 to N

phase = ["gas", "liquid", "solid"]
colors = ["r", "g", "b"]

plt.figure(figsize=(16, 8))
plt.suptitle("Energia totale", fontsize=25)

for i, phase_folder in enumerate(phase):
    media_column = []
    err_column = []

    file_path = "./" + phase_folder + "/output_etot.dat"
    with open(file_path, 'r') as file:
        for line in file:
            columns = line.split()

            if len(columns) >= 4:  # Ensures there are at least four columns
                second_value = columns[2].strip()
                fourth_value = columns[3].strip()

                if second_value and fourth_value:  # Skips empty strings
                    media_column.append(float(second_value))
                    err_column.append(float(fourth_value))

        plt.errorbar(np.arange(len(media_column)), media_column, yerr=err_column, color=colors[i], label=phase[i])

plt.title('Energia totale', fontsize=17, fontweight="bold")
plt.ylabel('$E^*/N$', fontsize=15)
plt.xlabel('Step number', fontsize=15)
plt.grid(True)
plt.legend()
plt.subplots_adjust(hspace=0.5)

plt.show()
```
