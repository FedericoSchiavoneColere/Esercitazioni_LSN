$f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$


```python
# target parameters of f(x)
a=-1.5
b=1.5


##

bs=32 #batch_size
epk=50 #epochs
n=5000 #numero punti
```

```python
import numpy as np

# Example data
np.random.seed(0)
x_train = np.random.uniform(a, b, n)
y_train = np.random.uniform(a, b, n)

x_valid, y_valid = np.meshgrid(np.linspace(a, b, 15), np.linspace(a, b, 15))

x_valid = x_valid.flatten()
y_valid = y_valid.flatten()
z_target = np.sin(x_valid**2 + y_valid**2)  # ideal (target) linear function

sigma = 0.15  # noise standard deviation
z_train = np.random.normal(np.sin(x_train**2 + y_train**2), sigma)
z_valid = np.random.normal(np.sin(x_valid**2 + y_valid**2), sigma)


# Convert the data to numpy arrays
x_train = np.array(x_train)
y_train = np.array(y_train)
z_train = np.array(z_train)
z_valid = np.array(z_valid)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Generate a meshgrid for x and y coordinates
x, y = np.meshgrid(np.linspace(a, b, 150), np.linspace(a, b, 150))

# Compute the corresponding z values
z_target_reshaped = np.sin(x**2 + y**2)

# Plot a basic wireframe of the target dataset.
ax.plot_wireframe(x, y, z_target_reshaped, rstride=10, cstride=10, label='Target')

# Plot the scatter points of the validation dataset.
ax.scatter(x_valid, y_valid, z_valid, marker='o', color='r', label='Validation')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

ax.view_init(elev=20, azim=45)  # Set initial viewpoint

plt.show()

```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from keras.layers import  Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects


# Define the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(34, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(34, activation='relu'),

    tf.keras.layers.Dense(1)
])



# Compile the model
#model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.compile(optimizer='adam', loss='mse')

# Combine x_train and y_train into a single input array
xy_train = np.column_stack((x_train, y_train))
```

```python
# get a summary of our composed model
model.summary()
```

```python
# Train the model and store the training history
history = model.fit(xy_train, z_train, 
                    validation_data=(np.column_stack((x_valid, y_valid)), z_valid), 
                    epochs=epk, batch_size=bs)
```

```python
# return weights and biases
#model.get_weights()
```

```python
# Concatenate x_valid and y_valid into a single input tensor
xy_valid = np.column_stack((x_valid, y_valid))

# Evaluate model
score = model.evaluate(xy_valid, z_valid, batch_size=bs, verbose=1)

# print performance
print('Test loss:', score)
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='best')
plt.show()
```

```python
x_new = np.linspace(a, b, num=100)
y_new = np.linspace(a, b, num=100)
x_mesh, y_mesh = np.meshgrid(x_new, y_new)
xy_new = np.column_stack((x_mesh.flatten(), y_mesh.flatten()))

# Predict the output for the new points
z_pred = model.predict(xy_new)
z_pred = z_pred.reshape(x_mesh.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Repeat the values of x_new and y_new to match the shape of z_pred
x_new_repeated = np.repeat(x_new, z_pred.shape[0])
y_new_repeated = np.tile(y_new, z_pred.shape[1])

# Plot the points
ax.scatter(x_new_repeated, y_new_repeated, z_pred.flatten(), c=z_pred.flatten(), cmap='cool', alpha=0.2, label='Points', s=5)

# Plot the mesh and the fit
ax.plot_surface(x_mesh, y_mesh, z_pred, cmap='viridis', alpha=0.9, label='Model')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Model Fit')
ax.view_init(elev=33, azim=40)

# Show the plot
plt.show()
```
