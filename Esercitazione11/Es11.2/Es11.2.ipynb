
$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-2,2]$.

2+2

```python
# target parameters of f(x)
a=3.
b=-2.
c=-3.
d=4.

##

bs=250 #batch_size
epk=500 #epochs
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-2, 2, 5000)
x_valid = np.random.uniform(-2, 2, 300)
x_valid.sort()
y_target = a*x_valid**3+ b*x_valid**2 + c*x_valid + c # ideal (target) linear function

sigma = 1.0 # noise standard deviation
y_train = np.random.normal(a*x_train**3+ b*x_train**2 + c*x_train + c, sigma) 
y_valid = np.random.normal(a*x_valid**3+ b*x_valid**2 + c*x_valid + c, sigma)
```

```python
# plot validation and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
#from tensorflow.keras.activations import elu



model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(4, input_shape=(1,), activation='tanh'),
    #tf.keras.layers.Dense(4, activation='elu'),
    #tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')
])


# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=bs, epochs=epk,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=bs, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=bs, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-2, 2, 100)
y_predicted = model.predict(x_predicted)
print(x_predicted.shape)
print(y_predicted.shape)
```
