```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as ss
```

```python
with open("unif1.txt") as unif1:
    dati_unif1 = np.fromstring(unif1.read(), sep=" ")
with open("unif2.txt") as unif2:
    dati_unif2 = np.fromstring(unif2.read(), sep=" ")
with open("unif10.txt") as unif10:
    dati_unif10 = np.fromstring(unif10.read(), sep=" ")
with open("unif100.txt") as unif100:
    dati_unif100 = np.fromstring(unif100.read(), sep=" ")
```

```python
with open("expo1.txt") as expo1:
    dati_expo1 = np.fromstring(expo1.read(), sep=" ")
with open("expo2.txt") as expo2:
    dati_expo2 = np.fromstring(expo2.read(), sep=" ")
with open("expo10.txt") as expo10:
    dati_expo10 = np.fromstring(expo10.read(), sep=" ")
with open("expo100.txt") as expo100:
    dati_expo100 = np.fromstring(expo100.read(), sep=" ")
```

```python
with open("cauchy_lorentz1.txt") as cauchy_lorentz1:
    dati_cauchy_lorentz1 = np.fromstring(cauchy_lorentz1.read(), sep=" ")
with open("cauchy_lorentz2.txt") as cauchy_lorentz2:
    dati_cauchy_lorentz2 = np.fromstring(cauchy_lorentz2.read(), sep=" ")
with open("cauchy_lorentz10.txt") as cauchy_lorentz10:
    dati_cauchy_lorentz10 = np.fromstring(cauchy_lorentz10.read(), sep=" ")
with open("cauchy_lorentz100.txt") as cauchy_lorentz100:
    dati_cauchy_lorentz100 = np.fromstring(cauchy_lorentz100.read(), sep=" ")
```

```python
bins = np.linspace(-0.5, 6.5, 100)
plt.hist(dati_unif1, bins, alpha=0.5, label='unif1', density=True)
plt.hist(dati_unif2, bins, alpha=0.5, label='unif2', density=True)
plt.hist(dati_unif10, bins, alpha=0.5, label='unif10', density=True)
plt.hist(dati_unif100, bins, alpha=0.5, label='unif100', density=True)
plt.legend(loc='upper right')
plt.show()
```

```python
bins = np.linspace(0, 6, 100)
plt.hist(dati_expo1, bins, alpha=0.5, label='expo1', density=True)
plt.hist(dati_expo2, bins, alpha=0.5, label='expo2', density=True)
plt.hist(dati_expo10, bins, alpha=0.5, label='expo10', density=True)
plt.hist(dati_expo100, bins, alpha=0.5, label='expo100', density=True)
plt.legend(loc='upper right')
plt.show()
```

```python
bins = np.linspace(0, 10, 100)
plt.hist(dati_cauchy_lorentz1, bins, alpha=0.5, label='cauchy_lorentz1', density=True)
plt.hist(dati_cauchy_lorentz2, bins, alpha=0.5, label='cauchy_lorentz2', density=True)
plt.hist(dati_cauchy_lorentz10, bins, alpha=0.5, label='cauchy_lorentz10', density=True)
plt.hist(dati_cauchy_lorentz100, bins, alpha=0.5, label='cauchy_lorentz100', density=True)
plt.legend(loc='upper right')
plt.show()
```
