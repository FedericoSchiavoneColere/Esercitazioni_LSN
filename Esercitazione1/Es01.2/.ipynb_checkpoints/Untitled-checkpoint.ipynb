```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as ss
```

```python
with open("expo.txt") as expo:
    dati_expo = np.fromstring(expo.read(), sep=" ")
```

```python
with open("cauchy_lorentz.txt") as CL:
    dati_CL = np.fromstring(CL.read(), sep=" ")
    
```

```python
P = ss.expon.fit(dati_expo)

rX = np.linspace(0,7,1000)
rP = ss.expon.pdf(rX, *P)

plt.hist(dati_expo, bins=200 ,density=True)
plt.plot(rX, rP);
plt.show();
```

```python
plt.hist(dati_CL,200,range=(-15, 15),density=True)
plt.show();
```
