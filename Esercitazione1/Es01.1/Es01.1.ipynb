```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
with open("medie.txt") as medie:
    dati_medie = np.fromstring(medie.read(), sep=" ")
```

```python
with open("errori.txt") as errori:
    dati_err = np.fromstring(errori.read(), sep=" ")
```

```python
plt.errorbar(np.arange(len(dati_medie))*1000, dati_medie -1/2 , yerr=dati_err)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
with open("medie_dev.txt") as medie_dev:
    dati_medie_dev = np.fromstring(medie_dev.read(), sep=" ")
```

```python
with open("errori_dev.txt") as errori_dev:
    dati_err_dev = np.fromstring(errori_dev.read(), sep=" ")
```

```python
fig = plt.figure()
```

```python
plt.errorbar(np.arange(len(dati_medie_dev))*1000, dati_medie_dev -1/12 , yerr=dati_err_dev)
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```
