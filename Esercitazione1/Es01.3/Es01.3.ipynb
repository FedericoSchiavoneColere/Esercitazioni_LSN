```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
with open("medie.txt") as medie:
    dati_medie = np.fromstring(medie.read(), sep=" ")
```

```python
with open("errori.txt") as errori:
    dati_err = np.fromstring(errori.read(), sep=" ")
```

```python
#fig = plt.figure()
```

```python
plt.errorbar(np.arange(len(dati_medie)), dati_medie, yerr=dati_err)
plt.xlabel('Cicli')
plt.ylabel('Pi-greco')
plt.grid(True)
plt.show()
```

```python
with open("cos1.txt") as cos1:
    dati_cos1 = np.fromstring(cos1.read(), sep=" ")
with open("cos2.txt") as cos2:
    dati_cos2 = np.fromstring(cos2.read(), sep=" ")
```

```python
plt.hist(dati_cos1,bins=100,alpha = 0.5,range=(-0.1,1.1),label='cos1',density=True)
plt.hist(dati_cos2,bins=100,alpha = 0.5,range=(-0.1,1.1),label='cos2',density=True)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
