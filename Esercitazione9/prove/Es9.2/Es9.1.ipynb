```python
import matplotlib.pyplot as plt
import numpy as np

# Read the points from a file
def read_points_from_file(filename):
    points = []
    with open(filename, 'r') as file:
        for line in file:
            x, y = line.strip().split(' ')
            points.append((float(x), float(y)))
    return points

# Read the points from a file
points = read_points_from_file('pos_iniz.txt')

# Create a figure and axis
fig, ax = plt.subplots()

# Set the aspect ratio to equal, ensuring a circular shape
ax.set_aspect('equal')

# Define the circle parameters
center = (0, 0)  # Center of the circle
radius = 2       # Radius of the circle

# Generate theta values from 0 to 2pi
theta = np.linspace(0, 2 * np.pi, 100)

# Calculate x and y coordinates for each theta value
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

# Plot the circle
ax.plot(x, y)

# Plot the points
ax.scatter(*zip(*points), label='Points')

# Connect the points with lines
ax.plot(*zip(*points), linestyle='--', color='red', label='Lines')

# Add labels to the points
for i, point in enumerate(points):
    ax.annotate(str(i + 1), point, textcoords="offset points", xytext=(5, 5), ha='center')

# Connect the last point to the first point
ax.plot([points[-1][0], points[0][0]], [points[-1][1], points[0][1]], linestyle='--', color='red')

# Set the title of the plot
ax.set_title('una configurazione iniziale')

# Set the legend position to upper right
ax.legend(loc='upper right')

# Show the plot
plt.show()
```
